#### Hypothesis Testing

```python
from scipy.stats import chi2_contingency
from scipy.stats import kstest, shapiro
```

```python
# Chi-Square Test 
def chi_square_test(df, columns, target):

    for column in columns:
        crosstab = pd.crosstab(df[target], df[column])
        chi2, p, dof, expected = chi2_contingency(crosstab)
        print(f'Chi-Square Test for {column} and {target}')
        print(f'Chi2: {chi2}')
        print(f'P-Value: {p}')
        print(f'Degrees of Freedom: {dof}')

        if p < 0.05:
            print(f'Reject Null Hypothesis :: There is a strong relationship between severity and {column} \n')
        else:
            print(f'Fail to Reject Null Hypothesis :: There is no relationship between severity and {column} \n')
```

```python
chi_square_test(cleandf, reqd_cols, target_column)
```

```python
def normality_test(data, alpha=0.05):
    statistic, pvalue = kstest(data, 'norm')
    print("Kolmogorov-Smirnov Test Statistic:", statistic)
    print('p-value:', pvalue)
    if pvalue > alpha:
        print("Sample looks Gaussian (fail to reject H0)")
    else:
        print("Sample does not look Gaussian (reject H0)")

for column in numeric_columns:
    print(f'Normality Test for {column}')
    normality_test(cleandf[column])
```

```python
sample = cleandf.sample(n=1000)

def shapiro_wilk_test(data, alpha = 0.05):
    
    statistic, pvalue = shapiro(data)
    print("Shapiro-Wilk Test Statistic:", statistic)
    print('p-value', pvalue)
    if pvalue > alpha: print("Sample looks Gaussian (fail to reject H0) \n")

    else: print("Sample does not look Gaussian (reject H0) \n")

for column in numeric_columns:
    print(f'Shapiro-Wilk Test for {column}')
    shapiro_wilk_test(sample[column])
```
